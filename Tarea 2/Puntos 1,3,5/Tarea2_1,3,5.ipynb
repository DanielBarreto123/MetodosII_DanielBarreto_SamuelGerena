```python
import numpy as np
import matplotlib.pyplot as plt

# ecuacion para x'=x
def differential_equation(x):
    return x

def runge_kutta1(f, x0, t0, tn, h):
    t_values = np.arange(t0, tn + h, h)
    x_values = []

    x = x0
    for t in t_values:
        x_values.append(x)
        k1 = h * f(x)
        x += k1

    return np.array(t_values), np.array(x_values)

def runge_kutta2(f, x0, t0, tn, h):
    t_values = np.arange(t0, tn + h, h)
    x_values = []

    x = x0
    for t in t_values:
        x_values.append(x)
        k1 = h * f(x)
        k2 = h * f(x + 0.5 * k1)
        x += k2

    return np.array(t_values), np.array(x_values)

def runge_kutta3(f, x0, t0, tn, h):
    t_values = np.arange(t0, tn + h, h)
    x_values = []

    x = x0
    for t in t_values:
        x_values.append(x)
        k1 = h * f(x)
        k2 = h * f(x + 0.5 * k1)
        k3 = h * f(x + 0.5 * k2)
        x += k3

    return np.array(t_values), np.array(x_values)

def runge_kutta4(f, x0, t0, tn, h):
    t_values = np.arange(t0, tn + h, h)
    x_values = []

    x = x0
    for t in t_values:
        x_values.append(x)
        k1 = h * f(x)
        k2 = h * f(x + 0.5 * k1)
        k3 = h * f(x + 0.5 * k2)
        k4 = h * f(x + k3)
        x += (k1 + 2*k2 + 2*k3 + k4) / 6

    return np.array(t_values), np.array(x_values)

# condiciones al azar
x0 = 1.0
t0 = 0.0
tn = 10
h = 0.5
def exponential_function(x):
    return np.exp(x)
x_values = np.linspace(0,10, 100)
y_values = exponential_function(x_values)
t_values1, x_values1 = runge_kutta1(differential_equation, x0, t0, tn, h)
t_values2, x_values2 = runge_kutta2(differential_equation, x0, t0, tn, h)
t_values3, x_values3 = runge_kutta3(differential_equation, x0, t0, tn, h)
t_values4, x_values4 = runge_kutta4(differential_equation, x0, t0, tn, h)
plt.plot(x_values, y_values, label='$f(x) = e^x$')
plt.plot(t_values1, x_values1, label='Runge-Kutta 1')
plt.plot(t_values2, x_values2, label='Runge-Kutta 2')
plt.plot(t_values3, x_values3, label='Runge-Kutta 3')
plt.plot(t_values4, x_values4, label='Runge-Kutta 4')


plt.legend()
plt.show()
```

Se observa que el método de RK3 tiende a sobreestimar los datos calculados, a diferencia de su predecesor RK2, el cual tiende a subestimarlos. A su vez, los métodos de orden inferior a RK4 muestran proximidad a la función de solución hasta un cierto punto, mientras que RK4 es el único que exhibe consistencia al siempre estar cerca de converger hacia la solución real.

```python
N = 100
t1 = np.linspace(2**0.5,5,N)
h1 = t1[1] - t1[0]
t2 = np.linspace(2**0.5,1,N)
h2 = t2[1] - t2[0]
def f(x,y):
    return (x**4 * y**2 - 2 * x**2 - 1) / x**3
```

```python
def GetEuler2(x,y,h):

    for i in range(1,len(x)):

        k1 = f(x[i-1],y[i-1])
        k2 = f(x[i-1]+h,y[i-1] + h*k1)

        y[i] = y[i-1] + 0.5*h*(k1+k2)
y1 = np.zeros_like(t1)
y1[0]=0
y2 = np.zeros_like(t1)
y2[0]=0
```

```python
GetEuler2(t1,y1,h1)
plt.plot(t1,y1)
GetEuler2(t2,y2,h2)
plt.plot(t2,y2)
```

```python
u0 = 1.0
t0 = 0.0
tn = 10
h = 0.1


def ff(u, q):
    return u**q

def GetEuler2_1(f, u0, t0, tn, h, q):
    tv= np.arange(t0, tn + h, h)
    uv= []
    u = u0
    for t in tv:
        uv.append(u)
        k1 = h * f(u, q)
        k2 = h * f(u + k1, q)
        u += 0.5 * (k1 + k2)

    return np.array(tv), np.array(uv)
```

```python
q=0
tt, uu = GetEuler2_1(ff, u0, t0, tn, h, q)
plt.plot(tt, uu, label=f'Solución numérica (n={q})')
plt.xlabel('Tiempo')
plt.ylabel('Valor de x')
plt.title(f'Ecuación Diferencial: x\' = x^{q}')
plt.legend()
plt.show()
```
