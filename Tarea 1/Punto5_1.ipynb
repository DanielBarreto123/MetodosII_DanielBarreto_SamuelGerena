```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate

def f(x,a):
    return a*x
def freal(x,a):
  return np.exp(a*x)

def GetEuler(a, u0, delta_t, num_steps):
    u = np.zeros(num_steps + 1)
    u[0] = u0
    for k in range(num_steps):
        u[k + 1] = u[k] * (1 + a * delta_t)
    return u
```

```python
N = 100
t = np.linspace(0,5,N)
```

```python
yreal = freal(t,-1)
```

```python
y1 = np.zeros_like(t)
y1[0] = 0
ya = GetEuler(-1,1,1.1,N-1)
yb = GetEuler(-1,1,1.5,N-1)
yc = GetEuler(-1,1,1.9,N-1)
plt.plot(t,yreal,color="k",label='real')
plt.plot(t,ya,color="b",label='1.1')
plt.plot(t,yb,color="r",label='1.5')
plt.plot(t,yc,color="g",label='1.9')
plt.legend()
```
